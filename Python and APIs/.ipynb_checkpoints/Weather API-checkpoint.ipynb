```python
def weather_data(x): #generates unique list of latitudes and longitudes to grab with citipy
    city_list = []
    rand_latlon = set()
    for i in range(x):
        lat = random.uniform(-90,90)
        while lat in rand_latlon:
            lat = random.uniform(-90,90)
        rand_latlon.add(lat)
        lon = random.uniform(-180,180)
        while lon in rand_latlon:
            lon = random.uniform(-180,180)
        rand_latlon.add(lon)
        city = citipy.nearest_city(lat, lon)
        city_name = city.city_name
        city_list.append(city_name)
    return city_list
```

```python
city_list = weather_data(50)
```

```python
weather_data = pd.DataFrame(columns = ['City','Country ID','Temperature','Latitude','Longitude','Humidity','Cloudiness','Wind Speed (mph)'],index=range(len(city_list)))
type(weather_data)
```
