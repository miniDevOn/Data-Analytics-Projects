```python
import tweepy
import json
from textblob import TextBlob
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import csv
```

```python
def get_file_contents(filename):
    """ Given a filename,
        return the contents of that file
    """
    try:
        with open(filename, 'r') as f:
            # It's assumed our file contains a single line,
            # with our API key
            return f.read().strip()
    except FileNotFoundError:
        print("'%s' file not found" % filename)
```

```python
consumer_key = get_file_contents('twitterConsumerKey.txt') #mXRZdblnQq2S3uXjjzEHIlZxK
consumer_secret = get_file_contents('twitterConsumerSecret.txt') #3XjwD8Y6HOdcbhduyuqztowU1IjAbO4dLK8MPoUaPzTVR4wEhK
access_token = get_file_contents('twitterAccessToken.txt') #860658468-ELyaxfXu8Z858D1Ku6w8mmvc9Yvj61hLqBeWFEUg
access_token_secret = get_file_contents('twitterAccessTokenSecret.txt') #P46Y1zJzgSJKedQr3GNZwrCnpd3OG54Wm85DDjnVcSWbN
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
```

```python
api = tweepy.API(auth,parser=tweepy.parsers.JSONParser())
```

```python
def add_data_to_df(name):
    tweet_df = pd.DataFrame(columns = ['Account Name','Tweet Text','Compound Score','Positive Score','Neutral Score','Negative Score','Date'])
    tweets = []
    sentiments = []
    text = []
    compound = []
    positive = []
    neutral = []
    negative = []
    dates = []
    names = []
    analyzer = SentimentIntensityAnalyzer()
    for i in range(5):
        tweets.append(api.user_timeline(name, page = i))
    for i in range(5):
        for x in range(20):
            sentiments.append(analyzer.polarity_scores(tweets[i][x]['text']))
            text.append(tweets[i][x]['text'])
            dates.append(tweets[i][x]['created_at'])
            names.append(tweets[i][x]['user']['name'])

    for i in range(100):
        compound.append(sentiments[i]['compound'])
        positive.append(sentiments[i]['pos'])
        neutral.append(sentiments[i]['neu'])
        negative.append(sentiments[i]['neg'])
    
    tweet_df['Account Name'] = pd.Series(names).values
    tweet_df['Tweet Text'] = pd.Series(text).values
    tweet_df['Date'] = pd.Series(dates).values
    tweet_df['Compound Score'] = pd.Series(compound).values
    tweet_df['Positive Score'] = pd.Series(positive).values
    tweet_df['Neutral Score'] = pd.Series(neutral).values
    tweet_df['Negative Score'] = pd.Series(negative).values
    
    return tweet_df
    #tweet_df.to_csv('%s.csv' % name, encoding = 'utf-8')
```

```python
bbc_df = add_data_to_df('BBCWorld')
bbc_df.to_csv('BBCWorld.csv')
cbs_df = add_data_to_df('CBSNews')
cbs_df.to_csv('CBSNews.csv')
cnn_df = add_data_to_df('CNN')
cnn_df.to_csv('CNN.csv')
fox_df = add_data_to_df('FoxNews')
fox_df.to_csv('FoxNews.csv')
nyt_df = add_data_to_df('nyTimes')
nyt_df.to_csv('NYTimes.csv')
```
