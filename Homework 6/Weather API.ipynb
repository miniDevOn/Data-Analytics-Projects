# Weather Py

### Observations
- It would appear that temperatures do peak near the equator, with temperatures at their highest in the southern hemisphere (where it is summer) and declining to its minimum in the northern hemisphere (where it is winter).
- Cities near the equator appear to have higher Humidity levels than average. Humidity and Cloudiness levels appear be higher in cities located in the northern hemisphere, with a large cluster of high Humidity locations near Latitude 60. This may be due to the typical weather changes associated with Winter.
- Wind Speed seems to be pretty constant despite changes in Latitude. The majority of measured Wind Speeds are below 10 mph, with an outlier speed of approximately 47 mph near Latitude 42.

```python
from citipy import citipy
import requests as req
import pandas as pd
import matplotlib.pyplot as plt
import json
import random
import csv
import datetime
```

```python
def weather_data(x): #generates unique list of latitudes and longitudes to grab with citipy
    city_list = []
    rand_latlon = set()
    for i in range(x):
        lat = random.uniform(-90,90)
        lon = random.uniform(-180,180)
        print(lat,lon)
        #while lat in rand_latlon:
         #   lat = random.uniform(-90,90)
        #while lon in rand_latlon:
         #   lon = random.uniform(-180,180)
        rand_latlon.add(lat)
        rand_latlon.add(lon)
        city = citipy.nearest_city(lat, lon)
        city_name = city.city_name
        if city_name not in city_list:
            city_list.append(city_name)

    return city_list
```

```python
city_list = weather_data(2000) """ In an effort to reduce the number of duplicate cities which are put into the list,
I needed to call a much higher number of random lat/lons to produce an adequate number of unique cities for use
in the analysis."""
```

# Perform API Calls, create DataFrame, and export CSV

```python
weather_data = pd.DataFrame(columns = ['City','Country ID','Temperature','Latitude','Longitude','Humidity','Cloudiness','Wind Speed (mph)','Requested URL'],index=range(len(city_list)))
api = '235e7cfc83c5d67891e91001012db21f'
url = 'http://api.openweathermap.org/data/2.5/weather?'
print('Beginning Data Retrieval')
print('-----------------------------------')
counter = 1
for i in range(len(city_list)):
    query_url = url + "appid=" + api + "&q=" + city_list[i] + '&units=imperial' # Build query URL, imperial units
    weather_response = req.get(query_url)
    weather_json = weather_response.json()

    if weather_json['cod'] != '404': #to eliminate city names that aren't found with Citipy
        print('City Number: ' + str(counter) + ' | City Name: ' + weather_json.get('name'))
        print('Requested URL: ' + str(query_url))
        print('-----------------------------------')
        counter += 1
        weather_data['City'][weather_data['City'].index[i]] = weather_json.get('name')
        weather_data['Country ID'][weather_data['Country ID'].index[i]] = weather_json['sys']['country']
        weather_data['Temperature'][weather_data['Temperature'].index[i]] = weather_json['main']['temp']
        weather_data['Latitude'][weather_data['Latitude'].index[i]] = weather_json.get('coord').get('lat')
        weather_data['Longitude'][weather_data['Longitude'].index[i]] = weather_json.get('coord').get('lon')
        weather_data['Humidity'][weather_data['Humidity'].index[i]] = weather_json.get('main').get('humidity')
        weather_data['Cloudiness'][weather_data['Cloudiness'].index[i]] = weather_json.get('clouds').get('all')
        weather_data['Wind Speed (mph)'][weather_data['Wind Speed (mph)'].index[i]] = weather_json.get('wind').get('speed')
        weather_data['Requested URL'][weather_data['Requested URL'].index[i]] = query_url                       

weather_data.to_csv('Weather Data.csv', encoding='utf-8',index=True)
```

```python
weather_data = weather_data.dropna() #Drop NaN values from DataFrame
weather_data
```

# Latitude vs Longitude Plot

```python
fig, ax = plt.subplots(figsize = (10,8),facecolor='lightgrey')
ax.set_ylim(-190,190)
ax.set_xlim(-95, 95)
plt.scatter(weather_data['Latitude'],weather_data['Longitude'],c='orange',marker='o',edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Latitude vs Longitude of randomly selected cities: '+ datetime.date.today().strftime("%B %d, %Y"))
plt.grid(linestyle='dotted',linewidth=1.5)
plt.savefig('Latitude vs Longitude')
plt.show()
```
